```python
#!/bin/bash

n=1 
while [ $n -le 5 ]; do           # -le  less then or equal 
    echo "Iteration number $n"
    ((n+=1))         # (( can perform arithmetic operations inside ))
done
```

```python
# random-exit.py script

#!/usr/bin/env python

import sys
import random

value=random.randint(0, 3)
print("Returning: " +str(value))
sys.exit(value)
```

```python
# retry.sh script

#!/bin/bash

n=1 
command=$1  # #  $1 is the first command argument passed in python --> sys.argv[1]

while ! $command && [ $n -le 5 ]; do
    sleep $n
    ((n=n+1))
    echo "Retry #$n"
done;
```



#### the random-exit.py is passed as an argument for retry.sh script 
    - random outputs a random exit value  
    - retry.sh check if that value is 0  --> ! $command
